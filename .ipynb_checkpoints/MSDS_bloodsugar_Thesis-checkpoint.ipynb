```python
#สร้าง function ให้แก้ค่า age (ติดปัญหา .item())
def fillage (df,age,cid,year,agelist):
    "fillage"
    for i in agelist:
        a = df[cid] == i
        b = df[age].isna()
        c = df[age].notnull()
        if (df[year][a&b] < df[year][a&c].min()).item:
            d = df[year][a&c].min() - df[year][a&b]
            df[age][df[cid]==i] = df[age][df[cid]==i].fillna(df[age][a&c].min()-d)
        else:
            e = df[year][a&b] - df[year][a&c].min()
            df[age][df[cid]==i] = df[age][df[cid]==i].fillna(df.age[a&c].min()+e)
```
